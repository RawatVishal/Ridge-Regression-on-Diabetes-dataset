# Separating dependent and independent features

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     x, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
##Model Training
```

```python
from sklearn.linear_model import Ridge
```

```python
l2 = Ridge()
```

```python
l2
```

```python
l2.fit(X_train,y_train)
```

```python
print(l2.coef_)
```

```python
print(l2.intercept_)
```

```python
## predicting for the test data 
```

```python
l2.pred = l2.predict(X_test)
```

```python
l2.pred
```

```python
## Assumption of ridge regression
```

```python
plt.scatter(y_test,l2.pred)
plt.xlabel("test truth data")
plt.ylabel("test predicted data")
```

```python
residuals = y_test-l2.pred
```

```python
residuals
```

```python
sns.displot(residuals,kind = 'kde') # it is left skewed
```

```python
plt.scatter(l2.pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,l2.pred))
print(mean_absolute_error(y_test,l2.pred))
print(np.sqrt(mean_squared_error(y_test,l2.pred)))
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,l2.pred)
print(score)
```

```python
adjusted_r_squared = 1 - (1-score)*(len(y_test)-1)/(len(y_test)-x.shape[1]-1)
```
